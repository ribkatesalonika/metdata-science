Jawaban no.3 
1) saya memilih menggunakan bar chart karena cocok digunakan saat ingin membandingkan nilai antar beberapa kategori untuk menunjukkan nilai terkecil dan terbesar dari suatu data.
2) 3 makanan penutup yang harus disingkirkan antara lain Kue Wajik,Pastel dan Puding Vanila karena ketiga makanan tersebut memiliki jumlah penjualan yang sangat sedikit dibanding 10 menu lainnya.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

nama_dessert = ('Donat','Pastel','Kue Coklat','Ice Cream','Puding Vanila','Brownies','Puding Strawberi','Puding Coklat',
                'Ice Cream Nutela','Kue Coklat-Keju','Kue Wajik','Kue Sus','Mochi')
jumlah_penjualan = (14,5,12,19,6,8,12,9,10,17,2,9,13)

df = pd.DataFrame({'Dessert':nama_dessert,'countofsale':jumlah_penjualan,})
df.sort_values(by='countofsale',inplace = True)

kor_x = np.arange(len(df))
warna = ['#0000FF' for _ in range(len(df))]
warna [0]= '#FF0000'
warna [1]= '#FF0000'
warna [2]= '#FF0000'
plt.figure(figsize=(13,9))

plt.bar(kor_x,df['countofsale'],tick_label = df['Dessert'], color = warna)
plt.xticks(rotation=90)
plt.ylabel('Jumlah penjualan')
plt.title('Penjualan Makanan Penutup Kafe Biru dalam 1 Minggu')
plt.show()
```
